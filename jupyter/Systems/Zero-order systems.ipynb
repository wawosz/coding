**Zero-order system** is the simplest example of a linear system defined by the ordinary differential equation (ODE) shown below:   

$a_0 y(t) + a_1 \dfrac{\delta y}{\delta t} + ... + a_n \dfrac{\delta^n y}{\delta t^{n}} = b_0 x(t) + b_1 \dfrac{\delta x}{\delta t} + ... + b_m \dfrac{\delta^m x}{\delta t^m}$

In zero-order system, the coefficients $n, m = 0$ which simplifies the ODE to it's trivial form:   

$y(t) = \dfrac{b_0}{a_0} x(t)$   

where $b_0/a_0$ is defined as static sensitivity (or gain) _K_.   

Let's define the zero-order system and plot it's responses to step and sinusoidal input functions.

```python
# standard imports
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# definition of the system where K is static sensitivity
def zero_order(x, K):
    output = K * x
    return output
```

### System's response to step input function
First, let's evaluate zero-order system's response to step input function modeled after piecewise function.    
Let's define and plot both input x(t) and the output y(t).

```python
# definition of the piecewise input function (step input function)
t = np.linspace (-5, 5, 100)
x = np.piecewise(t, [t<0, t>=0], [0, 1])
plt.plot(t, x)
plt.grid(False)
plt.title('Step input function x(t)')
plt.xlabel('Time (t)')
plt.ylabel('Input x(t)')
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# Plotting system's response to step input function
y = zero_order(x, 0.5)
plt.plot(t, y) 
    
plt.grid(False)
plt.title('Zero-order system response y(t)')
plt.xlabel('Time (t)')
plt.ylabel('Output y(t)')
plt.ylim(-0.1, 1.1)
plt.show()
```

### System's response to sinusoidal inputs
Next, let's evaluate zero-order system's response to sinusoidal input at frequency $f$.    
Let's define and plot a sinusoidal input and the output y(t).

```python
# definition of the sinusoidal input function x(t) with frequency 1 Hz
t = np.linspace (0, 5, 100)
f = 1
x = np.sin(2*np.pi*f*t)
plt.plot(t, x)
plt.grid(True)
plt.title('Sinusoidal input function x(t)')
plt.xlabel('Time (t)')
plt.ylabel('Input x(t)')
plt.ylim(-1.1, 1.1)
plt.show()
```

```python
# Plotting system's response to sinusoidal input function
y = zero_order(x, 0.5)
plt.plot(t, y) 
    
plt.grid(True)
plt.title('Zero-order system response y(t)')
plt.xlabel('Time (t)')
plt.ylabel('Output y(t)')
plt.ylim(-1.1, 1.1)
plt.show()
```

### Example of a zero-order system: Potentiometer-based sensor
In a real life, an example of zero-order system is a sensor based on potentiometer (variable resistor) as shown in the figure below. Irrespective of the frequency content of the input voltage, the output is merely $V_{out} = K \times V_{in}$, where static sensitivity $0 < K < 1$. 
![](zero_order_pots.png)

```python
# Example: potentiometer-based sensor response with K = 0.5
input_voltage = np.linspace(0, 1)
output_voltage = zero_order(input_voltage, 0.5)
plt.plot(input_voltage, output_voltage) 
    
plt.grid(True)
plt.title('Zero-order system response \n POTENTIOMETER-BASED SENSOR')
plt.xlabel('Input voltage (V)')
plt.ylabel('Output voltage (V)')
plt.show()
```

If you plot input voltage versus output voltage, the static sensitivity $K$ can be determined by slope of the line.

### Diode circuit model

To describe the electrical characteristics of diodes the following model has been proposed:

$i_D = i_0 (e^{\frac{q}{kT}V_D} - 1)$

where:

$i_D$ is current flowing across the diode

$i_0$ is the reverse saturation current (when diode is reverse-biased). $i_0 \approx 10^{-16}$A

$q$ is the electron charge

$K$ is Boltzmann's constant

$T$ is temperature in Kelvins

```python
import numpy as np
import matplotlib.pyplot as plt

i0 = 10**-16   # reverse saturation current in A
q = 1.60217662 * 10**-19   # electron charge in C
K = 1.38064852 * 10**-23   # Boltzmann constant in m2 kg s-2 K-1
T = 298.15   # temperature in K (25C)

vD = np.linspace(-1, 0.0001, num=100)
try:
    # iD = i0 * (np.exp((q/K * T) * vD) - 1)
    # simplified at room temperature
    iD = i0 * (np.exp(39*vD)-1)
except OverflowError:
    iD = float('inf')
plt.plot(vD + 0.7, iD)
plt.title("Simplified model of diode circuit")
plt.xlabel("Diode voltage, $V_D$ (V)")
plt.ylabel("Diode current, $i_D$ (A)")
plt.show()
```
